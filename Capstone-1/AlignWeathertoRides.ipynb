```python
import pandas as pd
import numpy as np
```

```python
# create dataframes for ride and weather dataset
ride_df = pd.read_csv('data/uberlyft_rides.csv')
weather_df = pd.read_csv('data/weather.csv')
```

```python
# convert epoch timestamp to datetime object
ride_df['time_stamp'] = pd.to_datetime(ride_df['time_stamp'], unit = 'ms')
weather_df['time_stamp'] = pd.to_datetime(weather_df['time_stamp'], unit = 's')

ride_df.head()
```

```python
# sort dataframes based on datetimes in ascending order
ride_df = ride_df.sort_values(by=['time_stamp'])
weather_df = weather_df.sort_values(by=['time_stamp'])

ride_df.info()
ride_df
```

```python
# rename loaction to source for merging purposes
weather_df.rename(columns={"location": "source"}, inplace=True)

weather_df.head()
```

```python
# replace empty rain values with 0
values = {'rain' : 0}
weather_df = weather_df.fillna(value=values)

weather_df.head()
```

```python
# merge weather dataframe with rides dataframe based on pickup location and nearest timestamp
comb_df = pd.merge_asof(ride_df, weather_df, on='time_stamp', by='source', tolerance=pd.Timedelta('1 hour'))

comb_df.head()
```

```python
# remove rows with no merged weather data and reset index
comb_df = comb_df.dropna().reset_index(drop=True)

comb_df.head()
```
