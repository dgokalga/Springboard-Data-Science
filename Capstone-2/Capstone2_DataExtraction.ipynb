# Giantbomb

```python
"""
Extracts review data from GiantBomb's API, and for each unique game, extracts game features: 
name,deck,description,developers,genres,franchises,platforms,publishers,characters,concepts,similar_games,themes
and concatenates game features to their respective game review record.

Input: offset, or page limit for each 100 game reviews

Returns list of dataframes, each containing 100 review records
"""


def gb_getdata(offset):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}
    r = requests.get("https://www.giantbomb.com/api/user_reviews/?api_key=7baa42265c0c5d6a4e4d4713c66b771833a6d827&format=JSON&field_list=date_added,deck,description,reviewer,score,site_detail_url,wikiObject&offset=" + str(offset), headers=headers)
    reviews = r.json()
    review = reviews['results'] 

    df = pd.DataFrame(review) # convert review json to dataframe
    df = df[df.wikiObject.notnull()].reset_index(drop=True)

    game_list = []
    
    # for each game from the 100 reviews, get game features for each game
    for url in df.site_detail_url.values: 
        time.sleep(5) # prevent API overloading
        guid = (url.split('/')[4]) # unique game id for each game
        r = requests.get("https://www.giantbomb.com/api/game/" + str(guid) + "/?api_key=7baa42265c0c5d6a4e4d4713c66b771833a6d827&format=JSON&field_list=name,deck,description,developers,genres,franchises,platforms,publishers,characters,concepts,similar_games,themes", headers=headers)
        game = r.json()
        game = game['results'] 
        game_list.append(game)

    game_df = pd.DataFrame(game_list) # convert game json to dataframe
 
    df = pd.concat([df, game_df], axis=1) # concatenate game features to their respective game features
    new_list.append(df)
    return new_list
```
